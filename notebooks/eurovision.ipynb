```python
import pandas as pd
import numpy as np
pd.options.display.max_columns=None
```

```python
df = pd.read_csv('../datos/eurovision.csv')
df.head()
```

```python
df = df[df['year'].isin([2022, 2021, 2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012, 2011,2010, 2009])]
```

```python
df['year'].unique()
```

```python
df = df[df['section'] =='grand-final']
```

```python
df.drop(['event_url','artist_url','section','image_url','country_emoji'],axis=1,inplace=True)
```

```python
df.shape
```

```python
df_2 = pd.read_csv('../datos/song_data.csv', encoding='ISO-8859-1')
df_2.head(1)
```

```python
df_2 = df_2[df_2['final_total_points'] >= 0]
df_2.head(1)
```

```python
df_2.drop(['semi_draw_position','selection','semi_place', 'semi_televote_points',
       'semi_jury_points', 'semi_total_points','qualified','release_date', 'key_change_10','age','direct_qualifier_10'],axis=1,inplace=True)
```

```python
df_2.head(1)
```

```python
def limpieza(elemento):
    return elemento.lower().replace('  ',' ').replace(' ','_')
```

```python
df['song'] = df['song'].apply(limpieza)
df_2['song_name'] = df_2['song_name'].apply(limpieza)
```

```python
df_union = df.merge(df_2,left_on='song',right_on='song_name')
```
